```python
# Source dataset (read-only)
base_input_path = os.path.join(path, "Dataset_BUSI_with_GT")

benign_src = os.path.join(base_input_path, "benign")
malignant_src = os.path.join(base_input_path, "malignant")
normal_src = os.path.join(base_input_path, "normal")
```

```python
base_output_path = "/content/BUSI_images"
```

```python
benign_dst = os.path.join(base_output_path, "benign")
malignant_dst = os.path.join(base_output_path, "malignant")
normal_dst = os.path.join(base_output_path, "normal")

os.makedirs(benign_dst, exist_ok=True)
os.makedirs(malignant_dst, exist_ok=True)
os.makedirs(normal_dst, exist_ok=True)
```

```python
def copy_images(src, dst):
    if not os.listdir(dst):   # avoid re-copying
        for img in os.listdir(src):
            if "mask" in img.lower():
                continue
            shutil.copy(
                os.path.join(src, img),
                os.path.join(dst, img)
            )

copy_images(benign_src, benign_dst)
copy_images(malignant_src, malignant_dst)
copy_images(normal_src, normal_dst)
```

```python
print("Benign images:", len(os.listdir(benign_dst)))
print("Malignant images:", len(os.listdir(malignant_dst)))
print("Normal images:", len(os.listdir(normal_dst)))
```
